```python
import numpy as np
from sklearn import datasets, linear_model
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
purpose_df = pd.read_csv('purposeofvisit.csv')
purpose_df.head()
```

```python
purpose_df1 = purpose_df.copy()
purpose_df1 = purpose_df1.drop(columns = ['vfr', 'holiday', 'general_business_purpose', 'mice'])
purpose_df1
```

```python
import re
import csv
```

```python
purpose_df2 = purpose_df1.copy()
purpose_df2['cor'] = purpose_df2['cor'].str.lower() 
purpose_df2['cor'] = purpose_df2['cor'].str.replace(r'[\n]', '')

purpose_df2['region'] = purpose_df2['region'].str.lower()
purpose_df2.head(7)
```

```python
purpose_df3 = purpose_df2[['leisure', 'business', 'others', 'not_stated']].apply(lambda x: x/100)
purpose_df2[['leisure', 'business', 'others', 'not_stated']] = purpose_df3
purpose_df2.head()
```

```python
purpose_df2.head(99)
```

```python
arrival_df = pd.read_csv('CHANGIAIRPORT1314.csv')
arrival_df.columns = ['Country','Arrivals13','Departures','Arrivals14','Departures']
arrival_df
```

```python
arrival_df = arrival_df.dropna().drop(columns = ['Departures'])
arrival_df['Country'] = arrival_df['Country'].str.lower()
arrival_df = arrival_df.iloc[1:]
arrival_df = arrival_df.replace({'united kingdom': 'uk', 'hong kong': 'hongkong'}) # to make common name between the 2 datasets
arrival_df
```

```python
def Arrivals13(x):
    for row in arrival_df.itertuples():
        if x in str(row.Country):
            return row.Arrivals13
    else:
        return 'N/A'
    
purpose_df2['Arrivals13'] = purpose_df2['cor'].apply(Arrivals13)
```

```python
def Arrivals14(x):
    for row in arrival_df.itertuples():
        if x in str(row.Country):
            return row.Arrivals14
    else:
        return 'N/A'
    
purpose_df2['Arrivals14'] = purpose_df2['cor'].apply(Arrivals14)
purpose_df2
```

```python
purpose_df2 = purpose_df2.groupby(['region'], as_index= False).agg('mean')
purpose_df2
# purpose_df2 = purpose_df2.drop([7,10,11,14,15,16], axis = 0)
```

```python
replacedNA = purpose_df2['Arrivals13'].replace('N/A', 0)
purpose_df2['Arrivals13'] = replacedNA
replaced = purpose_df2['Arrivals13'].replace(',', '')
purpose_df2['Arrivals13'] = replaced


purpose_df2.head(99)
```

```python
purpose_df2['leisure'] = purpose_df2['leisure'] * purpose_df2['Arrivals13']
purpose_df2['business'] = purpose_df2['business'] * purpose_df2['Arrivals13']
purpose_df2['others'] = purpose_df2['others'] * purpose_df2['Arrivals13']
purpose_df2['not_stated'] = purpose_df2['not_stated'] * purpose_df2['Arrivals13']

# This portion is the calculation to replace the values

purpose_df2.head(99)
```
